#Gaussiann Blur Filter:

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read an image
image_path = r".\Gaussian_Image.jpg"
image = cv2.imread(image_path)
```

```python
# Check if the image is loaded successfully
if image is None:
    print(f"Error: Unable to load the image at {image_path}")
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Apply a filter (e.g., Gaussian blur)
blurred_image = cv2.GaussianBlur(gray_image, (85, 85), 0)
```

```python

# Concatenate original and processed images horizontally
comparison_image = np.hstack((gray_image, blurred_image))
```

```python
plt.imshow(image, cmap='gray')
plt.title('Original')
plt.show()
plt.imshow(blurred_image, cmap='gray')
plt.title('Gaussian Blurred')
plt.show()
```
