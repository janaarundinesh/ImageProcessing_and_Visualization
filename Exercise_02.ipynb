```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read an image
image_path = r"E:\Universitat\Exams\ImageProcessing_and_visualization\Xray_Board.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Check if the image is loaded successfully
if image is None:
    print(f"Error: Unable to load the image at {image_path}")
```

```python
# Display the original image
plt.figure(figsize=(8, 8))
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
```

```python
# Define a simple 3x3 averaging filter/kernel
kernel3 = np.ones((3, 3), np.float32) / 9
kernel11 = np.ones((11, 11), np.float32) / 9
kernel21 = np.ones((21, 21), np.float32) / 9


# Apply the filter using cv2.filter2D
filtered_image1 = cv2.filter2D(image, -1, kernel3)
filtered_image2 = cv2.filter2D(image, -1, kernel11)
filtered_image3 = cv2.filter2D(image, -1, kernel21)
```

```python
# Display the filtered images
plt.subplot(131), plt.imshow(filtered_image1, cmap='gray'), plt.title('kernel 3x3')
plt.subplot(132), plt.imshow(filtered_image2, cmap='gray'), plt.title('kernel 11x11')
plt.subplot(133), plt.imshow(filtered_image3, cmap='gray'), plt.title('kernel 21x21')

# Adjust layout for better visualization
plt.tight_layout()

# Show the plots
plt.show()
```
